**Анализ существующих решений является важным этапом** в управлении проектами создания информационных систем.

#### **Теоретическая часть:**

**Анализ аналогов (конкурентный анализ)** - это систематическое исследование существующих на рынке решений, схожих с разрабатываемым продуктом, с целью выявления их преимуществ, недостатков, функциональных возможностей и определения конкурентной позиции собственного проекта.

**Основные цели анализа аналогов:**
- Изучение состояния рынка и существующих решений
- Выявление неудовлетворенных потребностей пользователей
- Определение уникальных характеристик разрабатываемого продукта
- Формирование конкурентных преимуществ

### **2. Этапы проведения анализа**

#### **Этап 1. Идентификация систем-аналогов**

Прежде всего необходимо определить критерии отбора систем для анализа:

**Критерии отбора:**
- **Функциональная схожесть** - решение выполняет схожие задачи
- **Целевая аудитория** - ориентировано на ту же группу пользователей
- **Масштаб** - сопоставимый уровень сложности и охват
- **Доступность информации** - возможность получить данные для анализа
- **Актуальность** - решение активно развивается и используется





**Источники поиска аналогов:**
- Специализированные каталоги ПО (Capterra, G2, Software Advice)
- Репозитории open-source проектов (GitHub, GitLab)
- Научные публикации и патентные базы
- Обзоры в профессиональных изданиях
- Поисковые системы и маркетплейсы приложений

**Типы аналогов для анализа:**
- **Прямые конкуренты** - решения с идентичным функционалом
- **Косвенные конкуренты** - решения, закрывающие ту же потребность иным способом
- **Потенциальные конкуренты** - решения, которые могут выйти на ваш рынок
- **Продукты-заменители** - альтернативные способы решения проблемы

#### **Этап 3. SWOT-анализ**

**SWOT-анализ** - стратегический метод анализа, оценивающий факторы внутренней и внешней среды организации или продукта.



**Компоненты SWOT:**

**S (Strengths - Сильные стороны)** - внутренние положительные характеристики:
- Уникальные функции
- Высокое качество
- Известность бренда
- Большая пользовательская база
- Техническое превосходство

**W (Weaknesses - Слабые стороны)** - внутренние отрицательные характеристики:
- Отсутствие важных функций
- Устаревшие технологии
- Проблемы с производительностью
- Сложный интерфейс
- Высокая стоимость

**O (Opportunities - Возможности)** - внешние благоприятные факторы:
- Растущий рынок
- Новые технологии
- Изменение потребностей пользователей
- Слабость конкурентов

**T (Threats - Угрозы)** - внешние неблагоприятные факторы:
- Появление новых конкурентов
- Изменение законодательства
- Технологические изменения
- Экономические факторы



**Практические рекомендации по SWOT-анализу:**
- Будьте объективны и реалистичны
- Используйте конкретные факты, а не общие фразы
- Ранжируйте факторы по значимости
- Анализируйте взаимосвязи между квадрантами
- Формулируйте стратегии на основе выявленных факторов

###### **5.2.1. Закон Хика (Hick's Law, 1952) - Время принятия решения**

**Историческая справка:** Уильям Эдмунд Хик и Рэй Хайман независимо обнаружили логарифмическую зависимость между количеством альтернатив и временем принятия решения, основываясь на теории информации Шеннона.


**Математическая формула:**
$RT = a + b \cdot \log_2(n + 1)$

где:
- $RT$ - время реакции (reaction time) в миллисекундах
- $a$ - общее время, не связанное с принятием решения (≈200 мс для визуального стимула)
- $b$ - эмпирическая константа, зависящая от сложности решения (≈150 мс для простых решений)
- $n$ - количество альтернатив выбора

**Почему это работает:** Мозг обрабатывает альтернативы методом бинарного поиска, последовательно исключая половину вариантов. Количество информации, необходимой для выбора из n альтернатив, равно $\log_2(n)$ бит согласно теории информации.



**Практическое применение для анализа интерфейсов:**





- Для меню с 12 пунктами (Jira):

  - $RT_{Jira} = 200 + 150 \cdot \log_2(12 + 1) = 200 + 150 \cdot 3.7 = 755$ мс



- Для меню с 5 пунктами (Trello):

  - $RT_{Trello} = 200 + 150 \cdot \log_2(5 + 1) = 200 + 150 \cdot 2.58 = 587$ мс

**Вывод для комиссии:** "Меню Jira требует на 29% больше времени для принятия решения согласно закону Хика, что подтверждено эмпирическими исследованиями (Хик, 1952; Хайман, 1953)."

###### **5.2.2. Закон Фиттса (Fitts's Law, 1954) - Время достижения цели**


**Историческая справка:** Пол Фиттс разработал модель человеческого движения, основанную на теории информации, первоначально для военных целей (предсказание точности пилотов).



**Математическая формула:**
$MT = a + b \cdot \log_2\left(\frac{2D}{W}\right)$

где:
- $MT$ - время движения (movement time)
- $D$ - расстояние от начальной точки до центра цели
- $W$ - ширина цели (допустимая погрешность)
- $a, b$ - эмпирические константы (для мыши: $a$ ≈ 50 мс, $b$ ≈ 150 мс)



**Индекс сложности (Index of Difficulty):**
$ID = \log_2\left(\frac{2D}{W}\right)$

Измеряется в битах информации, необходимых для уточнения позиции.


**Почему это работает:** Движение к цели представляет собой серию субдвижений с визуальной обратной связью. Каждое субдвижение уменьшает расстояние до цели примерно вдвое (оптимальная стратегия управления).





**Стандарты минимальных размеров интерактивных элементов:**
- **iOS Human Interface Guidelines:** минимум 44×44 пикселя
- **Material Design (Android):** минимум 48×48 dp
- **Web Content Accessibility Guidelines (WCAG):** минимум 44×44 CSS пикселя


**Пример расчета для кнопки "Сохранить":**

- Плохой дизайн (маленькая кнопка 20×20 px):
  - $ID_{small} = \log_2\left(\frac{2 \cdot 500}{20}\right) = \log_2(50) = 5.64$ бит

- Хороший дизайн (стандартная кнопка 48×48 px):
  - $ID_{standard} = \log_2\left(\frac{2 \cdot 500}{48}\right) = \log_2(20.8) = 4.38$ бит





**Время нажатия:**
 - $MT_{small} = 50 + 150 \cdot 5.64 = 896$ мс
 - $MT_{standard} = 50 + 150 \cdot 4.38 = 707$ мс



**Вывод для комиссии:** "Кнопка размером 20px требует на 27% больше времени для точного нажатия согласно закону Фиттса, что увеличивает количество промахов и снижает эффективность работы."

###### **5.2.3. Закон Миллера (Miller's Law, 1956) - Магическое число 7±2**


**Историческая справка:** Джордж Миллер изучал пределы человеческой способности обрабатывать информацию и обнаружил, что емкость кратковременной памяти ограничена.



**Теоретическая основа - Емкость канала:**

- $C = \log_2 N$

где:
- $C$ - емкость канала в битах
- $N$ - количество различимых альтернатив



**Современные исследования (Cowan, 2001; Luck & Vogel, 1997):** Реальная емкость рабочей памяти составляет $4 \pm 1$ элемента для сложных объектов. Число 7±2 актуально только для простых, одномерных стимулов (цифры, буквы).

**Формула когнитивной перегрузки:**
$P_{error} = 1 - e^{-k(n-C)}$ при $n > C$

где:
- $P_{error}$ - вероятность ошибки
- $n$ - количество элементов
- $C$ - емкость памяти (≈4 для сложных элементов)
- $k$ - константа сложности задачи (≈0.5)

**Практический пример:**




- Меню с 12 пунктами (превышение на 8 элементов):
  - $P_{error} = 1 - e^{-0.5 \cdot (12-4)} = 1 - e^{-4} = 0.98$ (98% вероятность ошибки или необходимости повторного просмотра)

- Меню с 5 пунктами (превышение на 1 элемент):
  - $P_{error} = 1 - e^{-0.5 \cdot (5-4)} = 1 - e^{-0.5} = 0.39$ (39% вероятность)

**Стратегия чанкинга (группировки):** Информация группируется в смысловые блоки, каждый из которых занимает одну ячейку памяти.

Пример: телефонный номер
- Плохо: 89991234567 (11 элементов)
- Хорошо: 8 (999) 123-45-67 (4 чанка)









**Вывод для комиссии:** "Интерфейс с 12 элементами меню превышает когнитивную емкость на 200%, что приводит к 98% вероятности необходимости повторного сканирования согласно модели Коуэна."

###### **5.2.4. Теория визуального внимания (Feature Integration Theory, Treisman & Gelade, 1980)**


**Основа теории:** Визуальная обработка происходит в два этапа:
1. **Предвнимательная стадия** - параллельная обработка базовых признаков
2. **Стадия фокусированного внимания** - последовательное связывание признаков

**Время визуального поиска:**
$T = a + b \cdot n$ (для поиска с участием внимания)
$T = a$ (для предвнимательного поиска)

где:
- $T$ - время поиска
- $n$ - количество отвлекающих элементов
- $a$ - базовое время (≈400 мс)
- $b$ - время на элемент (≈50 мс для сложного поиска)





**Базовые визуальные признаки (обрабатываются параллельно):**
- Цвет
- Ориентация
- Размер
- Движение



**Практическое применение:** Важные элементы должны отличаться по базовым признакам для мгновенного обнаружения.



**Пример анализа:**
"Кнопка 'Сохранить' в Jira не выделяется цветом среди 8 других кнопок, что увеличивает время поиска: $T = 400 + 50 \cdot 8 = 800$ мс вместо 400 мс при цветовом выделении."

###### **5.2.5. Гештальт-принципы восприятия (1920-е годы)**


**Теоретическая основа:** Мозг стремится к организации визуальной информации в целостные структуры согласно принципам:


**1. Закон близости (Proximity)**
Метрика группировки - коэффициент близости:
$K_{proximity} = \frac{d_{within}}{d_{between}}$

где:
- $d_{within}$ - среднее расстояние внутри группы
- $d_{between}$ - среднее расстояние между группами

Оптимальное значение: $K_{proximity} < 0.5$ (расстояние внутри группы минимум в 2 раза меньше)

**2. Закон схожести (Similarity)**
Метрика схожести по цвету (CIE ΔE*2000):
$\Delta E = \sqrt{(\Delta L^*)^2 + (\Delta a^*)^2 + (\Delta b^*)^2}$

Интерпретация:
- $\Delta E < 1.0$ - неразличимо для человека
- $\Delta E < 2.3$ - едва заметная разница
- $\Delta E < 5.0$ - заметная разница
- $\Delta E > 5.0$ - явная разница







**3. Закон общей области (Common Region)**
Элементы внутри замкнутой области воспринимаются как группа. Эффективность группировки:
$E_{region} = 1 - \frac{P_{cross}}{P_{total}}$

где:
- $P_{cross}$ - периметр пересечения областей
- $P_{total}$ - общий периметр областей

**Практический пример анализа группировки:**
"В интерфейсе Jira отступы между элементами одной группы составляют 16px, между группами - 20px, что дает коэффициент $K_{proximity} = 16/20 = 0.8 > 0.5$, следовательно, группировка визуально неочевидна."

##### **5.3. КОЛИЧЕСТВЕННЫЕ МЕТРИКИ UX**


###### **Метрики по стандарту ISO 9241-11:2018**



**Определение юзабилити:**
$Usability = f(Effectiveness, Efficiency, Satisfaction, Context)$


**1. Эффективность (Effectiveness):**
$Effectiveness = \frac{\sum_{i=1}^{N} \sum_{j=1}^{M} n_{ij} \times w_j}{N \times \sum_{j=1}^{M} w_j} \times 100\%$

где:
- $n_{ij}$ - результат задачи $j$ для пользователя $i$ (0 или 1)
- $w_j$ - вес задачи $j$
- $N$ - количество пользователей
- $M$ - количество задач

**2. Эффективность (Efficiency):**




$Time Efficiency = \frac{1}{N} \sum_{i=1}^{N} \sum_{j=1}^{M} \frac{n_{ij}}{t_{ij}}$



где $t_{ij}$ - время выполнения задачи $j$ пользователем $i$

**3. Относительная эффективность пользователя:**


$Relative User Efficiency = \frac{\sum_{i=1}^{N} \sum_{j=1}^{M} n_{ij} \times t_{ij}^{expert}}{\sum_{i=1}^{N} \sum_{j=1}^{M} n_{ij} \times t_{ij}^{user}} \times 100\%$


**Пример расчета для Jira:**
- Задача: создать и назначить задачу
- Время эксперта: 20 секунд
- Среднее время пользователя: 45 секунд
- Успешность: 78%

- $Effectiveness = 0.78 \times 100\% = 78\%$


- $Efficiency = \frac{0.78}{45} = 0.0173$ задач/сек


- $Relative Efficiency = \frac{20}{45} \times 100\% = 44.4\%$


**Вывод:** "Эффективность Jira составляет 44.4% от эталонной, что указывает на проблемы с юзабилити согласно ISO 9241-11."


##### **5.4. ОБЪЕКТИВНЫЕ МЕТРИКИ ВИЗУАЛЬНОГО ДИЗАЙНА**


###### **A. Визуальная иерархия и контраст**


**1. Формула контрастности WCAG 2.1:**


$CR = \frac{L_1 + 0.05}{L_2 + 0.05}$

где $L_1$ - относительная яркость более светлого цвета, $L_2$ - более темного

**Расчет относительной яркости:**
$L = 0.2126 \times R_{linear} + 0.7152 \times G_{linear} + 0.0722 \times B_{linear}$

**Линеаризация sRGB:**
$C_{linear} = \begin{cases}
\frac{C_{sRGB}}{12.92} & \text{если } C_{sRGB} \leq 0.03928 \\
\left(\frac{C_{sRGB} + 0.055}{1.055}\right)^{2.4} & \text{иначе}
\end{cases}$

**Требования WCAG 2.1:**
- Обычный текст: $CR \geq 4.5:1$
- Крупный текст (14pt жирный или 18pt обычный): $CR \geq 3:1$
- Уровень AAA: $CR \geq 7:1$

**2. Визуальный вес элемента:**
$W_{visual} = S \times C \times P \times M$

где:
- $S$ - относительный размер (площадь элемента / площадь экрана)
- $C$ - контраст с фоном (нормализованный 0-1)
- $P$ - позиционный вес (центр = 1, периферия = 0.6)
- $M$ - движение (статичный = 1, анимированный = 1.5)

###### **B. Типографические метрики**


**1. Оптимальная длина строки (Dyson & Haselgrove, 2001):**



Скорость чтения в зависимости от длины строки:
$Reading Speed = -0.002 \times (CPL - 66)^2 + 250$

где CPL - количество символов в строке

Максимум при CPL = 66 символов (250 слов/мин)

**2. Индекс читабельности Флеша-Кинкейда:**



$FKGL = 0.39 \times \frac{words}{sentences} + 11.8 \times \frac{syllables}{words} - 15.59$

Интерпретация:
- FKGL < 8: легко читается большинством
- FKGL 8-12: средняя сложность
- FKGL > 12: сложный текст





**3. Межстрочный интервал:**



$Line Height = Font Size \times k$

где $k$ оптимально:
- Для основного текста: $1.5 \leq k \leq 1.7$
- Для заголовков: $1.2 \leq k \leq 1.3$
- Для мобильных устройств: $k \geq 1.6$



##### **5.5. КОГНИТИВНАЯ НАГРУЗКА - ИЗМЕРЕНИЕ**



###### **Модель когнитивной нагрузки (Sweller, 1988)**

**Общая когнитивная нагрузка:**
$CL_{total} = CL_{intrinsic} + CL_{extraneous} + CL_{germane}$

При этом должно выполняться:
$CL_{total} \leq WM_{capacity}$

где $WM_{capacity}$ ≈ 4 единицы (Cowan, 2001)

**Измерение когнитивной нагрузки - NASA-TLX:**



$TLX = \frac{\sum_{i=1}^{6} rating_i \times weight_i}{\sum_{i=1}^{6} weight_i}$

Шкалы оценки (0-100):
1. Mental Demand (умственная нагрузка)
2. Physical Demand (физическая нагрузка)
3. Temporal Demand (временная нагрузка)
4. Performance (производительность)
5. Effort (усилия)
6. Frustration (фрустрация)


**Индекс визуальной сложности (Rosenholtz et al., 2007):**
$VC = \alpha \cdot FC + \beta \cdot SC$

где:
- $FC$ - feature congestion (перегруженность признаками)
- $SC$ - subband entropy (энтропия поддиапазонов)
- $\alpha, \beta$ - весовые коэффициенты

**Расчет перегруженности признаками:**
$FC = \sigma_{color} + \sigma_{luminance} + \sigma_{orientation}$

##### **5.6. МЕТРИКИ ОШИБОК И ВОССТАНОВЛЕНИЯ**



###### **Классификация ошибок (Norman, 1983)**


**1. Промахи (Slips):** правильное намерение, неправильное выполнение

**2. Ляпсусы (Lapses):** ошибки памяти

**3. Ошибки (Mistakes):** неправильное намерение







**Метрика частоты ошибок:**

- $Error Rate = \frac{Total Errors}{Total Attempts} \times 100\%$

**Метрика восстановления после ошибок:**

- $Recovery Rate = \frac{Recovered Errors}{Total Errors} \times 100\%$

**Время восстановления:**
- $Recovery Time = t_{complete} - t_{error} - t_{normal}$

##### **5.7. ЭМОЦИОНАЛЬНЫЙ ДИЗАЙН - ИЗМЕРЕНИЕ**



###### **Модель эмоций PAD (Mehrabian & Russell, 1974)**

Эмоциональное состояние описывается вектором:
$Emotion = (P, A, D)$

где:
- $P$ - Pleasure (удовольствие): от -1 до +1
- $A$ - Arousal (возбуждение): от -1 до +1
- $D$ - Dominance (доминирование): от -1 до +1

**Целевые значения для продуктивных приложений:**
- $P = 0.6$ (умеренно приятный)
- $A = 0.3$ (слегка стимулирующий)
- $D = 0.7$ (высокий контроль пользователя)

###### **Эстетическая привлекательность (Ngo et al., 2003)**


**Метрика визуального баланса:**

- $Balance = 1 - \frac{|W_L - W_R|}{W_L + W_R}$

где $W_L$, $W_R$ - визуальные веса левой и правой частей



**Метрика симметрии:**

- $Symmetry = 1 - \frac{\sum|p_i - p_i'|}{n \times max\_distance}$

где $p_i$, $p_i'$ - соответствующие точки относительно оси симметрии

##### **5.8. СТАТИСТИЧЕСКАЯ ВАЛИДАЦИЯ РЕЗУЛЬТАТОВ**



###### **Определение размера выборки для юзабилити-тестирования**


**Формула Nielsen (2000) для обнаружения проблем:**
$P = 1 - (1 - p)^n$

где:
- $P$ - вероятность обнаружения проблемы
- $p$ - доля пользователей с проблемой (≈0.31 для типичных интерфейсов)
- $n$ - количество участников

Для 85% проблем: $n = 5$
Для 95% проблем: $n = 10$


###### **Статистическая значимость различий**



**t-тест для сравнения времени выполнения задач:**
$t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$

**Размер эффекта Коэна:**
$d = \frac{\bar{x}_1 - \bar{x}_2}{s_{pooled}}$

Интерпретация:
- $d < 0.2$ - незначительный эффект
- $0.2 \leq d < 0.5$ - малый эффект
- $0.5 \leq d < 0.8$ - средний эффект
- $d \geq 0.8$ - большой эффект

**Анализ системы Jira по объективным метрикам:**




**1. Закон Хика - время выбора в меню:**
- Количество пунктов: 12

- **Вывод:** соответствует минимальным требованиям AA

##### **5.11. Контрольный чеклист**






Перед защитой убедитесь, что для каждой проблемы UX/UI у вас есть:

- **Конкретная метрика или формула** (закон Хика, Фиттса, и т.д.)
- **Численное значение** (время в мс, проценты, баллы)
- **Сравнение с эталоном или стандартом** (WCAG, ISO, HIG)
- **Ссылка на научное исследование** (автор, год)
- **Расчет или измерение** (формула с подставленными значениями)
- **Интерпретация результата** (что означает для пользователя)
- **Визуальное доказательство** (скриншот, схема, график)
- **Статистическая значимость** (при наличии выборки)



Это позволит вам уверенно обосновать свои выводы перед комиссией и доказать, что ваш анализ базируется на научных принципах и объективных измерениях, а не на субъективных предпочтениях.

- **A/B тестирование** - сравнение вариантов интерфейса
- **Анализ метрик** - изучение данных об использовании
- **Обзор отзывов пользователей** - анализ обратной связи





**Инструменты для UX/UI анализа:**

#### **Этап 6. Формулирование конкурентных преимуществ**

**Конкурентное преимущество** - характеристика продукта, отличающая его от конкурентов и создающая дополнительную ценность для пользователей.


**Виды конкурентных преимуществ:**



-  **1. Функциональные:**
      - Уникальные возможности
      - Более полный функционал
      - Лучшая реализация стандартных функций

-  **2. Технологические:**
    - Использование передовых технологий
    - Лучшая производительность
    - Более высокая надежность

-  **3. Пользовательские:**
    - Более удобный интерфейс
    - Лучший пользовательский опыт
    - Персонализация

-  **4. Экономические:**
    - Более низкая цена
    - Лучшее соотношение цена/качество
    - Гибкая модель монетизации

-  **5. Стратегические:**
    - Фокус на специфичной нише
    - Адаптация под особые потребности
    - Локализация



**Формулировка уникального торгового предложения (УТП):**

- УТП должно отвечать на три вопроса:
  1. Что вы предлагаете?
  2. Чем это отличается от конкурентов?
  3. Почему это важно для пользователя?



**Формула УТП:**



```
[Продукт] помогает [целевой аудитории] [решить проблему/получить выгоду],
в отличие от [конкуренты], благодаря [уникальная характеристика].
```

#### **5. Выявленные проблемы существующих решений**

На основе проведенного анализа выявлены следующие ключевые проблемы:

**1. Дихотомия сложности:**
- Jira: избыточная сложность для малых и средних команд
- Trello/GitHub Projects: недостаточный функционал для полноценного управления проектами
- **Проблема:** отсутствие решения с оптимальным балансом простоты и функциональности

**2. Зависимость от интернета:**
- Все рассмотренные коммерческие решения требуют постоянного онлайн-подключения
- **Проблема:** невозможность работы в условиях нестабильного интернета или повышенных требований к безопасности

**3. Слабая интеграция с инструментами тестирования:**
- Существующие системы не имеют глубокой интеграции с frameworks тестирования
- **Проблема:** разрозненность процессов разработки и тестирования

**4. Стоимость:**
- Полнофункциональные решения (Jira, Asana) дороги для малых команд
- Бесплатные решения имеют существенные ограничения
- **Проблема:** высокий порог входа для стартапов и малых команд

**5. Кастомизация:**
- Коммерческие решения имеют ограничения в кастомизации
- Open-source решения требуют высоких технических компетенций
- **Проблема:** сложность адаптации под специфические потребности команды

**Влияние:** Пользователи с огранниченым бюджетом не могут позволить себе использовать эти сервисы

**Распространенность:** Carto, ArcGIS

### 8.2. Принципиальные отличия

**Описание:** Все рассмотренные аналоги - это готовые продукты для визуализации и анализа пространственных данных. Их архитектура и интерфейс скрывают детали реализации.

**Влияние:** Разработчки, аналитики данных не могут получить ответы на вопросы об влиянии тех или инных технологий на производительность: "При каких объемах данных одна СУБД имеет преимущество над другой", "При каких объемах данных лучше использовать Kappa подход, а при каких лямбда", "Какой прирост производительности даст использование оконных функций", "Какой прирост производительности даст бинаризация данных" и др.

**Распространенность:** На всех аналогах

## 9. КОНКУРЕНТНЫЕ ПРЕИМУЩЕСТВА РАЗРАБАТЫВАЕМОГО РЕШЕНИЯ

### 9.1. Объективное сравнение трёх колоночных СУБД



**Описание:** Фреймворк позволяет запустить один и тот же набор запросов и визуализаций на ClickHouse, DuckDB и TimescaleDB, автоматически собирая метрики времени выполнения и FPS.

**Обоснование:** Это позволяет дать понимание пользователю о выполе СУБД

**Ценность для пользователя:** Понимание выбора СУБД от сценария пользователя

**Отличие от конкурентов:** У аналогов нет сравнения

### 9.2. Встроенные метрики производительности

**Описание:** В дашборде отображаются FPS, время перерисовки при зуме, скорость передачи данных.

**Обоснование:** Позволяет количественно оценить, как изменение уровня детализации, добавление количества объектов влияет на плавность.

**Ценность:** Разработчики видят порог, после которого карта начинает лагать

**Отличие от конкурентов:** Ни один аналог не предоставляет таких метрик в реальном времени.

### 9.3. Экспериментальное сравнение Lambda и Kappa архитектур

**Описание:** Реализованы два режима:  
- Lambda: исторические данные в DuckDB + потоковые в ClickHouse.  
- Kappa: все данные (история и поток) через Kafka + ClickHouse

**Обоснование:** Позволяет выбрать архитектуру в зависимости от требований к свежести данных и затратам на инфраструктуру.

**Ценность:** Технический архитектор может принять обоснованное решение.

**Отличие от конкурентов:** Аналоги поддерживают только один из подходов.

### 9.4. Открытый воспроизводимый бенчмарк

**Описание:** Весь код тестов, скрипты генерации данных и конфигураций выложен в открытый репозиторий.

**Обоснование:** Позволяет любому заинтересованному лицу повторить эксперимент на своих данных или дополнить тесты.

**Ценность:** Поможет в научных целях (эмпрический анализ).
